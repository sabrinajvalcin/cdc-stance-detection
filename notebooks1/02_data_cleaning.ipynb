# Data Cleaning

Hello team!
This notebook performs the first preprocessing stage.

Steps:
1. Load the raw dataset.
2. Import the cleaning functions from `scripts/preprocessing.py`.
3. Apply the cleaning pipeline.
4. Save:
   - `clean_periods.csv` (original + cleaned text)
   - `clean_text_only.csv` (only cleaned text)

```python
# Required libraries
from pathlib import Path
import pandas as pd
import sys

# Project root directory
ROOT_DIR = Path.cwd().parent

# Allows Python to import files from the "scripts" folder
sys.path.append(str(ROOT_DIR))

# Import the preprocessing function from scripts/preprocessing.py
from scripts.cleaning import clean_comments_dataframe
```

```python
# Input file (raw data)
RAW_PERIODS_FILE = ROOT_DIR / "data" / "raw" / "periods" / "all_periods.csv"

# Output file with original and cleaned text
CLEAN_PERIODS_FILE = ROOT_DIR / "data" / "clean" / "clean_periods.csv"

# Output file with only the cleaned text
CLEAN_TEXT_ONLY_FILE = ROOT_DIR / "data" / "clean" / "clean_text_only.csv"

# Load the dataset
df = pd.read_csv(RAW_PERIODS_FILE)
```

```python
# Apply the cleaning pipeline from preprocessing.py
clean_df = clean_comments_dataframe(df, text_column="text")

# Create the output folder if it does not exist
CLEAN_PERIODS_FILE.parent.mkdir(parents=True, exist_ok=True)

# Save the dataset with both original and cleaned text
clean_df.to_csv(CLEAN_PERIODS_FILE, index=False, encoding="utf-8-sig")

# Save a dataset containing only the cleaned text
clean_df[["clean_text"]].to_csv(CLEAN_TEXT_ONLY_FILE, index=False, encoding="utf-8-sig")

clean_df.head()
```
